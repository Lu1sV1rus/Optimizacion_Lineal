```python

import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
import numpy as np
#1.3
def graficar(x,y):
  a = np.arange(0,80)
  f1 = 2*a
  f2 = (80-a)/4
  f3 = (40-0.9*a)/0.8
  plt.plot(a,f1)
  plt.plot(f2,a)
  plt.plot(a,f3)
  plt.plot(x,y, marker="o", color="red")
  plt.xlabel('Eje X')
  plt.ylabel('Eje Y')
  plt.title('Grafica')
  plt.grid()
  plt.show()

def uti(x,y):
  u=50*x+20*y
  return u

B = np.array([
      [2,-1],
      [1,4],
      [0.9,0.8]])

D=np.array([
      [-1,0,0],
      [0,-1,0],
      [0,0,-1]])
b=np.array(
      [0,80,40]
      )
restric=np.array([0,0,0,0,0])
XSol=np.array([0,0,0,0,0])
puntos=[]
M=np.array([[],[],[]])
A=np.insert(B,B.shape[1],D,1)
print(A,"=",b)
mm=A.T
m=len(A)
n=len(mm)
for i in range(n-2):
  for jj in range(i+1,n-1):
    for j in range(jj+1,n):
        W=np.insert(M, M.shape[1], A.T[i], 1)
        W=np.insert(W, W.shape[1], A.T[jj],1)
        W=np.insert(W, W.shape[1], A.T[j],1)
        if np.linalg.det(W)!=0:
          B1=np.linalg.inv(W)
          X=B1.dot(b)
          cont=0
          XSol=np.array([0.0,0,0,0,0])
          XSol[i]=round(X[0], 4)
          XSol[jj]=round(X[1], 4)
          XSol[j]=round(X[2], 4)
          for k in range(n):
            if XSol[k]>=restric[k]:
              cont=cont+1
          if cont==5:
              print("---------------")
              graficar(XSol[0],XSol[1])
              print("X=",XSol)
              print(W)
              print("Utilidad=",uti(XSol[0],XSol[1]))
              puntos.append(XSol[0])
              puntos.append(XSol[1])
              puntos.append(XSol[2])
        W=np.array([[],[],[]])
        B1=np.array([[],[],[]])
        X=np.array([])
        XSol=np.array([0.0,0,0,0,0])
```
